## 1. Crie um array 2D `shape(10,10)`  e realize operações de indexação e fatiamento para obter o seguinte resultado.

```bash
[[ 0.  8.  0.  0.  0.  0.  0.  0. -8.  0.]
 [ 0.  8.  0.  2.  2.  2.  2.  2. -8.  2.]
 [ 5.  5.  5.  5.  5.  5.  5.  5. -8.  0.]
 [ 0.  8.  0.  0.  0.  0.  0.  0. -8.  0.]
 [ 0.  8.  0.  2.  2.  2.  2.  2. -8.  2.]
 [ 0.  8.  0.  0.  0.  0.  0.  0. -8.  0.]
 [ 0.  8.  0.  0.  0.  0.  0.  0. -8.  0.]
 [ 0.  8.  0.  2.  2.  2.  2.  2. -8.  2.]
 [ 0.  8.  0.  0.  0.  0.  0.  0. -8.  0.]
 [ 0.  8.  0.  0.  0.  0.  0.  0. -8.  0.]]
 ```

````python
# criando um array numpy com valores no shape 10x10 com zeros (valor que mais se repete) 
x = np.zeros((10,10), dtype=float)

# inserindo os valores corretos nos lugares corretos usando os conceitos de indexação e fatiamento:

# penultima coluna é toda composta por -8:
x[:,-2] = -8

# primeira linha:
x[0,:][1] = 8
x[0,:][-2] = -8

# segunda linha:
x[1,:][1] = 8
x[1,:][3:8] = 2
x[1,:][-1] = 2

# terceira linha:
x[2,:][0:8] = 5

# linhas que são exatamente iguais a outras linhas já ajustadas: 
x[3,:] = x[0,:] # a linha três é igual a linha zero
x[4,:] = x[1,:] # a linha quatro é igual a linha um
x[5,:] = x[0,:] # a linha cinco é igual a linha zero
x[6,:] = x[0,:] # a linha seis é igual a linha zero
x[7,:] = x[1,:] # a linha sete ém igual a linha um
x[8,:] = x[0,:] # a linha oito é igual a linha zero
x[9,:] = x[0,:] # a linha nove é igual a linha 0

print(x)
````

## 2. Crie um array 2D com a sequência de números ímpares no intervalo 0 a 100 e faça as seguintes operações:
- Média das linhas
- Média das colunas
- Soma dos elementos da última linha
- Soma dos elementos da última coluna

**Resultado Esperado:**
```bash
Média das linhas: [ 5. 15. 25. 35. 45. 55. 65. 75. 85. 95.]
Média das colunas: [46. 48. 50. 52. 54.]

Soma dos elementos da última linha: 475
Soma dos elementos da última coluna: 540
```

````python
# criando um array numpy com 50 elementos do tipo float no shape 10 por 5 apenas com números ímpares
impares = np.array([i for i in range(1, 100, 2)], dtype=float).reshape(10,5) 

media_linhas = impares.sum(axis = 1)/5          # soma todas as linhas e divide pelo numero total de linhas para obter a média das linhas
media_colunas = impares.sum(axis = 0)/10        # somas todas as colunas e divide pelo numero total de colunas para obter a média das colunas
soma_ultima_linha = int(impares[9,:].sum())     # soma dos elementos da última linha, transformado para inteiro para ter o resultado sem ponto flutuante como o esperado pelo enunciado
soma_ultima_coluna = int(impares[:,-1:].sum())  # soma dos elementos da última coluna, transformado para inteiro para ter o resultado sem ponto flutuante como o esperado pelo enunciado

print('Média das linhas:', media_linhas)
print('Média das colunas:', media_colunas)
print('Sooma dos elementos da última linha:', soma_ultima_linha)
print('Sooma dos elementos da última coluna:', soma_ultima_coluna)
````

## 3. Particione verticalmente ao meio o array `a`, armazenando a parte superior em `b` e inferior em `c`. Depois, tranforme os números negativos em positivos do array `c` e atribua `0` para os valores negativos do array `b`. Ao final das transformações, concatene `b` com `c` e forneça um gráfico de dispersão com os dados da matriz transposta (conforme está abaixo).
```python
a=np.array([[ 1.02809624,  0.31011686, -0.24478542, -0.43531192,  0.30234452, -1.5811674],
 [-0.0029375,   1.26637201,  0.14090036,  0.40795145, -0.57373286,  0.43089205],
 [ 1.76380711, -0.80405178,  1.16717726, -0.77396777,  1.89063668,  1.88322279],
 [ 1.26010925, -0.81197676, -0.90776823,  1.27519403, -0.29176333,  1.21126804],
 [ 0.43983984, -0.41200537,  0.85670881, -0.40561811, -0.44112721,  0.48759066],
 [ 0.21434576, -0.28789424, -0.25646132,  2.30080928, -0.34601413, -1.15924328]])
```
**Resultado Esperado**



````python
a = np.array([[ 1.02809624,  0.31011686, -0.24478542, -0.43531192,  0.30234452, -1.5811674],    # linha 0
            [-0.0029375,   1.26637201,  0.14090036,  0.40795145, -0.57373286,  0.43089205],   # linha 1
            [ 1.76380711, -0.80405178,  1.16717726, -0.77396777,  1.89063668,  1.88322279],   # linha 2
            [ 1.26010925, -0.81197676, -0.90776823,  1.27519403, -0.29176333,  1.21126804],   # linha 3
            [ 0.43983984, -0.41200537,  0.85670881, -0.40561811, -0.44112721,  0.48759066],   # linha 4
            [ 0.21434576, -0.28789424, -0.25646132,  2.30080928, -0.34601413, -1.15924328]])  # linha 5
            # são 6 linhas no total

# particionando verticalmente o array a passando a parte superior (3 primeiras linhas) para o b e a parte inferior (3 últimas linhas) para o arry c
b = a[0:3, : ] # b recebe as linhas 0, 1 e 2 
c = a[3:6, : ] # c recebe as linhas 3, 4, 5

# transformar os numeros negativos em positivos em c
c = np.abs(c)

# atribui zero para os valores negativos em b
b = np.maximum(b, 0) # o valor mais baixo possível será 0, todos os valores abaixo de zero serão transformados em zero

# juntando as matrizes b e c numa só matriz
matriz = np.concatenate([b,c])

# se eu faço matriz.transpose() como pede o enunciado o resultado sai diferente do esperado
matriz_transposta = matriz.T

# plotando o gráfico da matriz
plt.title('Dados da Matriz Transposta') # atribuindo um título ao gráfico
plt.scatter( matriz_transposta, matriz_transposta)

plt.show()
````

## 4. Forneça um único DataFrame a partir dos dados contidos em `alugueis_info.csv` e `alugueis_valores.csv`. Depois de corrigir os dados, salve o DataFrame corrigido em um arquivo chamado `alugueis.csv` para reusar nas próximas questões.

_Obs: Para quem usar laços nesta questão, terá um desconto de 25% na nota._


**Resultado Esperado do DataFrame:**
```bash
<class 'pandas.core.frame.DataFrame'>
Int64Index: 6073 entries, 0 to 6079
Data columns (total 13 columns):
 #   Column           Non-Null Count  Dtype  
---  ------           --------------  -----  
 0   Cidade           6073 non-null   bool   
 1   Area             6073 non-null   float64
 2   Quartos          6073 non-null   int64  
 3   Banheiros        6073 non-null   int64  
 4   Vagas Garagem    6073 non-null   int64  
 5   Andar            6073 non-null   int64  
 6   Animal           6073 non-null   bool   
 7   Mobiliado        6073 non-null   bool   
 8   Condominio       6073 non-null   float64

 12  Total            6073 non-null   float64

```
**Os 5 primeiros registros do DataFrame:**


````python
# lendo os arquivos
alugueis_info = pd.read_csv('alugueis_info.csv')
alugueis_valores = pd.read_csv('alugueis_valores.csv')

# concateno os dois csv em um só chamado alugueis
alugueis = pd.merge(alugueis_info, alugueis_valores, how = 'inner')

# retiro a coluna Unnamed pois o próprio pandas faz a indexação
alugueis = alugueis.drop('Unnamed: 0', axis=1) 

alugueis['Cidade'] = alugueis['Cidade'].astype(bool) # arrumando o tipo da coluna Cidade
alugueis['Andar'] = alugueis['Andar'].str.replace('-', '0') # tratando os símbolos - 

# a saida tipo da coluna Andar era int32 então fiz a conversão para int64 para o resultado ficar exatamente igual ao solicitado pelo professor
# REFENRECIA: https://stackoverflow.com/questions/46699280/how-to-convert-python-int-into-numpy-int64
alugueis['Andar'] = np.dtype('int64').type(alugueis['Andar'])

# mudar os tipos que tem R$ para float: Condominio, Aluguel, IPTU, Seguro Incendio, Total, desta forma, a saída ficará exatamente igual a esperada.
# Para isso farei alguns tratamentos dos dados:

# Tratando todos os casos de inconscistencia para transformar a coluna Condominio em float assim como na saida esperada
alugueis['Condominio'] = alugueis['Condominio'].apply(lambda x: x.replace('R$',''))
alugueis['Condominio'] = alugueis['Condominio'].apply(lambda x: x.replace(',',''))
alugueis['Condominio'] = alugueis['Condominio'].apply(lambda x: x.replace('Sem info','0'))
alugueis['Condominio'] = alugueis['Condominio'].apply(lambda x: x.replace('Incluso','0')).astype(float)

# Tratando todos os casos de inconscistencia para transformar a coluna Aluguel em float assim como na saida esperada
alugueis['Aluguel'] = alugueis['Aluguel'].apply(lambda x: x.replace('R$',''))
alugueis['Aluguel'] = alugueis['Aluguel'].apply(lambda x: x.replace(',','')).astype(float)     

# Tratando todos os casos de inconscistencia para transformar a coluna IPTU em float assim como na saida esperada
alugueis['IPTU'] = alugueis['IPTU'].apply(lambda x: x.replace('R$',''))
alugueis['IPTU'] = alugueis['IPTU'].apply(lambda x: x.replace(',',''))
alugueis['IPTU'] = alugueis['IPTU'].apply(lambda x: x.replace('Incluso','0')).astype(float)    

# Tratando todos os casos de inconscistencia para transformar a coluna Seguro Incendio em float assim como na saida esperada
alugueis['Seguro Incendio'] = alugueis['Seguro Incendio'].apply(lambda x: x.replace('R$',''))
alugueis['Seguro Incendio'] = alugueis['Seguro Incendio'].apply(lambda x: x.replace(',','')).astype(float)

# Tratando todos os casos de inconscistencia para transformar a coluna Total em float assim como na saida esperada
alugueis['Total'] = alugueis['Total'].apply(lambda x: x.replace('R$',''))
alugueis['Total'] = alugueis['Total'].apply(lambda x: x.replace(',','')).astype(float)

# depois de juntar os 2 csvs e tratar os dados necessários salvando alugueis em um csv
alugueis.to_csv

alugueis.info()
alugueis.head()
````

## 6. Para a base de dados `alugueis.csv`, apresente um gráfico de barras com a média do valor dos aluguéis relativo ao número de quartos para os imóveis com e sem mobilia. 

**Resultado Esperado:**


````python
# filtrando apenas as colunas que serão utilizadas nesta questao
aux = alugueis[['Mobiliado', 'Aluguel', 'Quartos']] 

# SEM MOBILIA
imoveis_sem_mobilia = aux.loc[aux['Mobiliado'] == False] # filtro somente os imoveis sem mobilia 
# calculo a meia do valor dos alugueis relativo ao numero de quartos definindo duas casas decimais após o ponto
dict_sem = dict(imoveis_sem_mobilia.groupby(['Quartos'])['Aluguel'].mean().round(2))

# COM MOBILIA
imoveis_com_mobilia = aux.loc[aux['Mobiliado'] == True] # filtro somente os imoveis com mobilia 
# calculo da media do valor dos alugueis relativo ao numero de quartos definindo duas casas decimais após o ponto
dict_com_mobilia = dict(imoveis_com_mobilia.groupby(['Quartos'])['Aluguel'].mean().round(2))

# iniciando o gráfico com as informações corretas
fig, ax = plt.subplots(figsize=(10, 6))

# para os dados com mobilia
x_dados = list(dict_com_mobilia.keys()) # as labels serão as achaves do dict
y_dados = list(dict_com_mobilia.values())

ax.bar(np.array(x_dados)-0.1, y_dados, align='center', width = 0.4)

for i in range(len(x_dados)):
    ax.annotate(y_dados[i],
                xy = (x_dados[i], y_dados[i]),
                xycoords ='data',
                xytext = (x_dados[i], y_dados[i]),
                textcoords = 'data',
                horizontalalignment = 'center',
                verticalalignment = 'bottom'
               )

# para os dados sem mobilia
x_dados = list(dict_sem.keys())
y_dados = list(dict_sem.values())
ax.bar(np.array(x_dados)+0.1, y_dados, align='edge', width = 0.4)

for i in range(len(x_dados)):
    ax.annotate(y_dados[i],
                xy = (x_dados[i], y_dados[i]),
                xycoords ='data',
                xytext = (x_dados[i], y_dados[i]),
                textcoords = 'data',
                horizontalalignment = 'center',
                verticalalignment = 'bottom'
               )

# declarando as labels do gráfico
plt.title('Média do valor dos aluguéis')
plt.xlabel('Número de Quartos')
plt.ylabel('Valor em Reais')
# declarando a legenda 
plt.legend(('Com mobilia', 'Sem mobilia')) 
plt.xticks(x_dados) # seto a "label" do x. Referencia: https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/bar_stacked.html#sphx-glr-gallery-lines-bars-and-markers-bar-stacked-py
plt.show()

# Nota para o professor: nesta questão faltou apenas posicionar melhor as barras, eu tentei de várias formas mas sempre gerava algum erro. Infelizmente, não consegui!
````

## 7. Para a base de dados `alugueis.csv`, apresente dois gráficos `boxplot`  (`Axes(1,2)`) do valor de `IPTU`, sendo um para os 3 maiores e outro para os 3 menores alugueis em média relativo a quantidade de banheiros.

_Dica: nesta questão calcula-se a média do aluguel por quantidade benheiros para saber quem são os 3 maiores e menores aluguéis. Depois, plota-se o IPTU para cada um deles._

**Resultado Esperado:**



````python
# encontrando a média dos aluguéis relativos aos banheiros truncado em duas casas decimais
media = pd.Series(alugueis.groupby(['Banheiros'])['Aluguel'].mean().round(2)).sort_values(ascending =False)

# pego os "maiores banheiros"
maiores_qtd_banheiros = media.nlargest(3) # https://pandas.pydata.org/docs/reference/api/pandas.Series.nlargest.html
# pego os "menores banheiros"
menores_qtd_banheiros = media.nsmallest(3) # https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nsmallest.html

# encontro o iptu dos alugueis que tem 1, 2 e 3 banheiros
menor1 = alugueis['IPTU'].loc[alugueis['Banheiros'] == menores_qtd_banheiros.index[0]]
menor2 = alugueis['IPTU'].loc[alugueis['Banheiros'] == menores_qtd_banheiros.index[1]]
menor3 = alugueis['IPTU'].loc[alugueis['Banheiros'] == menores_qtd_banheiros.index[2]]

# encontro o iptu dos alugueis que tem 10, 8 e 7 banheiros
maior1 = alugueis['IPTU'].loc[alugueis['Banheiros'] == maiores_qtd_banheiros.index[0]]
maior2 = alugueis['IPTU'].loc[alugueis['Banheiros'] == maiores_qtd_banheiros.index[1]]
maior3 = alugueis['IPTU'].loc[alugueis['Banheiros'] == maiores_qtd_banheiros.index[2]]

fig, ax = plt.subplots(1, 2, figsize=(10, 6))
plt.title('IPTU dos 3 Maiores e Menores Aluguéis Médio por Quantidade de Banheiros')

# # IPTU dos 3 maiores
dados = [maior1, maior2, maior3] # valores
dados_l = [maiores_qtd_banheiros.index[0], maiores_qtd_banheiros.index[1], maiores_qtd_banheiros.index[2]] # banheiros

ax[0].boxplot(dados, labels=dados_l, showfliers=False, showmeans=True)
ax[0].set_title('IPTU dos 3 Maiores', fontsize = 14)
ax[0].set_xlabel('Quantidade de Banheiros')
ax[0].set_ylabel('Valor em Reais')

# # IPTU dos 3 menores
dados = [menor1, menor2, menor3] # valores
dados_l = [menores_qtd_banheiros.index[0], menores_qtd_banheiros.index[1], menores_qtd_banheiros.index[2]] # banheiros

ax[1].boxplot(dados, labels=dados_l, showfliers=False, showmeans=True)
ax[1].set_title('IPTU dos 3 Menores', fontsize = 14)
ax[1].set_xlabel('Quantidade de Banheiros')
ax[1].set_ylabel('Valor em Reais')

plt.show()
````
